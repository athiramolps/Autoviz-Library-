# Generate visualizations

```python
%matplotlib inline

from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()

# Generate visualizations
AV.AutoViz("/content/World City Populations 2023.csv")
```

#Setting a Target Variable


```python
AV = AutoViz_Class()
dftc = AV.AutoViz(
    filename='',
    sep='' ,
    depVar='growthRate',
    dfte=df,
    header=0,
    verbose=1,
    lowess=False,
    chart_format='png',
    max_rows_analyzed=778,
    max_cols_analyzed=8
)
```
